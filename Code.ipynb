```python
cols = ['0.01%_ETH', '0.05%_ETH', '0.3%_ETH', '0.05%_ETHUSDT', '0.3%_ETHUSDT']
#find the difference between both prices 
#how it is capitalized to make full use of fees in price discrepancies and does volume increase
```

```python
dates = []
date_indexes = []
prices_df['Dates_to_note'] = 0
for i in range(len(prices_df)): 
    max_row = 0 
    max_col = ''
    for col in cols: 
        difference = abs(prices_df.loc[i, 'Binance_Price'] - prices_df.loc[i, col])  
        prices_df.loc[i, col+'_difference'] = difference 
        if ((difference/ prices_df.loc[i, 'Binance_Price'])*100) > 5:
            dates.append(prices_df.loc[i, 'date'])
            date_indexes.append(i)
            prices_df.loc[i, 'Dates_to_note'] = 1
prices_df
```

# Uniswap Price Volatility

```python
import numpy as np
vol_df = pd.DataFrame()
cols = ['0.01%_ETH','0.05%_ETH','0.3%_ETH','0.3%_ETHUSDT','0.05%_ETHUSDT']

for col in cols: 
    prices = prices_df[col]     
    window = 5  # trading days in rolling window
    days_per_year = 365  # trading days per year
    ann_factor = days_per_year / window
    log_return = np.log(prices).diff()
    # Var Swap (returns are not demeaned)
    real_var = np.square(log_return).rolling(window).sum() * ann_factor
    real_vol = np.sqrt(real_var)
    vol_df[col + '_volatility'] = real_vol*100
vol_df
```

```python
plt.figure(figsize=(10,5))

plt.plot(prices_df.date, vol_df['0.01%_ETH_volatility'], label = '0.01%_ETHUSDC_volatility')
plt.plot(prices_df.date, vol_df['0.05%_ETH_volatility'], label = '0.05%_ETHUSDC_volatility')
plt.plot(prices_df.date, vol_df['0.3%_ETH_volatility'], label = '0.3%_ETHUSDC_volatility')
plt.plot(prices_df.date, vol_df['0.3%_ETHUSDT_volatility'], label = '0.3%_ETHUSDT_volatility')
plt.plot(prices_df.date, vol_df['0.05%_ETHUSDT_volatility'], label = '0.05%_ETHUSDT_volatility')
plt.legend()
plt.show()
```
